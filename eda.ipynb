```python
import pandas as pd
```

```python
df1 = pd.read_csv("data/pop_2014_2023.csv", dtype='string', usecols=['UF', 'Municipio', 'popMun', "AnoBase", "codMun", "popUF"])
df1.tail()
```

```python
mun = pd.read_csv("data/df_mun_2024.csv", dtype='string', sep="\t")
mun.columns = ["UF", "codUF", "codMun", "Municipio", "popMun"]
mun.head()
```

```python
mun['codMun'] = mun['codUF'] + mun['codMun']
mun.head()
```

```python
crap = [f'({i})' for i in range(1, 33)]
for i in crap:
    mun['popMun'] = mun['popMun'].str.replace(i, '')

mun.head()
```

```python
mun['popMun'] = mun['popMun'].str.strip().str.replace('.','').astype(int)
mun.head()
```

```python
mun.to_csv("data/df_mun_2024.csv", index=False)
```

```python
uf = mun.groupby("UF")['popMun'].sum().reset_index()
uf.columns = ["UF", "popUF"]
uf.head()
```

```python
pop = pd.merge(mun, uf, on='UF')
pop.sample(5)
```

```python
pop['AnoBase'] = 2024
```

```python
pd.concat([df1, pop.drop('codUF', axis=1)], axis=0).to_csv('data/pop_2014_2024.csv', index=False)
```
